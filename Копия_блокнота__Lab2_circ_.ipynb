### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def add_front(self, item):
        self.items.insert(0, item)

    def add_rear(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None

    def remove_rear(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None


def sort_book_titles(input_file, output_file):
    deck1 = Deque()
    deck2 = Deque()

    with open(input_file, 'r') as file:
        for line in file:
            # Add each line to deck1
            deck1.add_rear(line.strip())
            print(f"Added to deck1: {line.strip()}")

    # Sort the lines using deck2
    while not deck1.is_empty():
        current_title = deck1.remove_front()
        print(f"Sorting: {current_title}")

        deck2_front = None
        # Find the correct position to insert the title in deck2
        while not deck2.is_empty():
            deck2_front = deck2.remove_rear()

            if deck2_front is not None and deck2_front[0] < current_title[0]:
                deck1.add_rear(deck2_front)
                print(f"Moved to deck1: {deck2_front}")

            elif deck2_front is not None and deck2_front[0] >= current_title[0]:
                deck2.add_rear(deck2_front)
                print(f"Returned to deck2: {deck2_front}")
                break
            else:
                break

        deck2.add_rear(current_title)
        print(f"Moved to deck2: {current_title}")

    # Save the sorted titles to the output file
    with open(output_file, 'w') as output:
        print("sorted titles: ")
        while not deck2.is_empty():
            sorted_title = deck2.remove_rear()
            print(sorted_title)
            output.write(sorted_title + '\n')


if __name__ == "__main__":
    input_filename = "bookTitleSource.txt"
    output_filename = "bookTitleSorted.txt"

    sort_book_titles(input_filename, output_filename)
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

def check_parentheses_balance(file_path):
    stack = Stack()

    with open(file_path, 'r') as file:
        line_number = 1

        for line in file:
            for char_index, char in enumerate(line, start=1):
                if char in '({[':
                    stack.push((char, line_number, char_index))
                elif char in ')}]':
                    if stack.is_empty():
                        print(f"Unmatched closing parenthesis '{char}' at line {line_number}, position {char_index}.")
                    else:
                        open_char, open_line, open_pos = stack.pop()
                        if not ((open_char == '(' and char == ')') or
                                (open_char == '{' and char == '}') or
                                (open_char == '[' and char == ']')):
                            print(f"Mismatched parenthesis '{open_char}' at line {open_line}, position {open_pos}.")

            line_number += 1

    while not stack.is_empty():
        open_char, open_line, open_pos = stack.pop()
        print(f"Unmatched opening parenthesis '{open_char}' at line {open_line}, position {open_pos}.")


if __name__ == "__main__":
    file_path = "algorithm.txt"
    check_parentheses_balance(file_path)
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
class SymbolStack:
    def __init__(self):
        self.numbers = []
        self.letters = []
        self.remaining = []

    def process_symbols(self, symbols):
        for symbol in symbols:
            if symbol.isdigit():
                self.numbers.append(symbol)
            elif symbol.isalpha():
                self.letters.append(symbol)
            else:
                self.remaining.append(symbol)

    def print_categorized_symbols(self):
        print("Numbers:", ''.join(self.numbers))
        print("Letters:", ''.join(self.letters))
        print("Remaining:", ''.join(self.remaining))

def categorize_symbols(file_path):
    symbol_stack = SymbolStack()

    with open(file_path, 'r') as file:
        for line in file:
            symbols = line.strip()
            symbol_stack.process_symbols(symbols)

    symbol_stack.print_categorized_symbols()

if __name__ == "__main__":
    file_path = "symbols.txt"  
    categorize_symbols(file_path)
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def add_front(self, item):
        self.items.insert(0, item)

    def add_rear(self, item):
        self.items.append(item)

    def remove_front(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None

    def remove_rear(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

def categorize_integers(input_file, output_file):
    negative_integers = Deque()
    positive_integers = Deque()

    # Read integers from input file and categorize them
    with open(input_file, 'r') as file:
        for line in file:
            integers = map(int, line.strip().split())
            for integer in integers:
                if integer < 0:
                    negative_integers.add_rear(integer)
                else:
                    positive_integers.add_rear(integer)

    # Write categorized integers to output file
    with open(output_file, 'w') as file:
        file.write("Negative Numbers:\n")
        while not negative_integers.is_empty():
            file.write(str(negative_integers.remove_front()) + "\n")

        file.write("Positive Numbers:\n")
        while not positive_integers.is_empty():
            file.write(str(positive_integers.remove_front()) + "\n")

if __name__ == "__main__":
    input_filename = "integers.txt"
    output_filename = "categorized_integers.txt"

    categorize_integers(input_filename, output_filename)
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

def reverse_lines(input_file, output_file):
    line_stack = Stack()

    # Read lines from the input file and push them onto the stack
    with open(input_file, 'r') as file:
        for line in file:
            line_stack.push(line.strip())

    # Write reversed lines to the output file
    with open(output_file, 'w') as file:
        while not line_stack.is_empty():
            file.write(line_stack.pop() + '\n')

if __name__ == "__main__":
    input_filename = "source.txt"  
    output_filename = "reversed_lines.txt"

    reverse_lines(input_filename, output_filename)
```

Стеки:

Стеки следуют принципу «Последним пришел — первым ушел» (LIFO).
Элементы добавляются (передвигаются) и удаляются (выталкиваются) с одного и того же конца (обычно сверху).
Последний элемент, помещенный в стек, извлекается первым.
Стеки обычно используются для управления вызовами функций, отслеживания истории выполнения и обработки функций отмены.
Дек:

Деки поддерживают операции на обоих концах, позволяя добавлять или удалять элементы как спереди, так и сзади.
Деки могут использоваться как стек (LIFO) или очередь (FIFO) в зависимости от требований.
Элементы можно добавлять или удалять спереди (слева) или сзади (справа) дека.
Деки представляют собой универсальные структуры данных и полезны в сценариях, где элементы необходимо эффективно добавлять или удалять с обоих концов.
В предоставленных скриптах Python мы использовали простую реализацию стека и дека, чтобы продемонстрировать их использование. Вот краткое описание того, как они применялись:

Пример стека:

Изменен порядок строк в текстовом файле с использованием стека.
Каждая строка из файла помещалась в стек, а затем строки извлекались из стека, чтобы записать их в обратном порядке.
Пример дека:

Классифицированные целые числа (отрицательные и положительные) из текстового файла с использованием двух деков.
Каждое целое число добавлялось либо в начало, либо в конец соответствующей дека, а затем классифицированные целые числа записывались в новый файл.
В обоих примерах выбор между использованием стека или дека был сделан исходя из конкретных требований задачи. Стеки использовались, когда требовался доступ только к последнему элементу, а деки использовались, когда требовалась большая гибкость в добавлении или удалении элементов с обоих концов.
